## Descripción del proyecto

Este proyecto busca analizar datos de COVID-19 a nivel mundial utilizando diferentes datasets.
El objetivo principal es comprender la distribución de casos confirmados, muertes y recuperaciones
por país y continente, y preparar la información para futuras etapas de Machine Learning.

Los datos provienen de distintas fuentes:
- **country_wise_latest**: Resumen de casos por país.
- **covid_19_clean_complete**: Evolución diaria de casos por país.
- **usa_county_wise**: Casos específicos por condado en Estados Unidos.
- **worldometer_data**: Datos agregados incluyendo población, tests y casos por millón.

El proyecto se desarrolla bajo la metodología **CRISP-DM**, y esta fase corresponde a la **Comprensión del Negocio**.


## Objetivos del proyecto

1. Identificar los países con mayor número de casos y muertes.
2. Analizar la distribución de casos por continente.
3. Explorar la evolución de los casos en el tiempo por país.
4. Preparar los datos para futuras fases de Machine Learning (predicción de casos futuros o análisis por región).
