## Descripcin del proyecto

Este proyecto busca analizar datos de COVID-19 a nivel mundial utilizando diferentes datasets.
El objetivo principal es comprender la distribucin de casos confirmados, muertes y recuperaciones
por pas y continente, y preparar la informacin para futuras etapas de Machine Learning.

Los datos provienen de distintas fuentes:
- **country_wise_latest**: Resumen de casos por pas.
- **covid_19_clean_complete**: Evolucin diaria de casos por pas.
- **usa_county_wise**: Casos especficos por condado en Estados Unidos.
- **worldometer_data**: Datos agregados incluyendo poblacin, tests y casos por milln.

El proyecto se desarrolla bajo la metodologa **CRISP-DM**, y esta fase corresponde a la **Comprensin del Negocio**.
