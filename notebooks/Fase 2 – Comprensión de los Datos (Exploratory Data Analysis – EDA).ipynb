# Estadistica descriptiva

```python
# Función para revisar dataset
def explore_dataset(df, name):
    print(f"===== Dataset: {name} =====")
    print("Dimensiones:", df.shape)
    print("Columnas:", df.columns.tolist())
    print("Valores nulos por columna:\n", df.isnull().sum())
    print("Filas duplicadas:", df.duplicated().sum())
    display(df.head(5))
    display(df.describe(include='all').transpose())

# Revisar datasets
explore_dataset(covid_df, "COVID-19 Clean Complete")
explore_dataset(usa_df, "USA County Wise")
explore_dataset(worldometer_df, "Worldometer")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 10 países con más casos confirmados
top_countries = covid_df.groupby("Country/Region")["Confirmed"].max().sort_values(ascending=False)[:10]
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="viridis")
plt.title("Top 10 países con más casos confirmados")
plt.ylabel("Casos confirmados")
plt.xticks(rotation=45)
plt.show()

# Distribución de casos por continente (worldometer)
continent_cases = worldometer_df.groupby("Continent")["TotalCases"].sum()
plt.figure(figsize=(8,8))
plt.pie(continent_cases, labels=continent_cases.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribución de casos COVID-19 por continente")
plt.show()
```
